# Getting Started with Amadeusz Agents

This notebook demonstrates basic setup and usage of the project.

```python
# Import required libraries
import os
from dotenv import load_dotenv
from openai import OpenAI

# Load environment variables
load_dotenv()
```

## Setup OpenAI Client

```python
# Initialize OpenAI client
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# Verify connection
print("OpenAI client initialized successfully!")
```

## Simple Example

```python
# Example: Simple completion
response = client.chat.completions.create(
    model="gpt-4",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "What is Python?"}
    ]
)

print(response.choices[0].message.content)
```

## Using Project Modules

```python
# Import from the agents package
from agents.example import simple_completion

# Use the helper function
result = simple_completion("Tell me a fun fact about Python programming!")
print(result)
```
